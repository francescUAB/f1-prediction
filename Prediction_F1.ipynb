```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.utils.class_weight import compute_class_weight


results = []
circuit=0
accuracy_summary = []
valores=0


for i, dataframe in enumerate(dataframes_circuits):
    # Filtrar y preparar datos
    df_filtered = dataframe[dataframe['target'] <= 20].dropna()
    X = df_filtered.drop(columns=['dob', 'date', 'driverref','circuitref', 'target','raceid'])
    y = df_filtered['target']

    min_samples = 30
    min_samples_per_class = 6
    
    # if len(y) < min_samples or len(np.unique(y)) < 2 or any(y.value_counts() < min_samples_per_class):
    #     print(f"No hay suficientes datos para Circuit_{i+1}. Saltando este circuito.")
    #     continue

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)


    if len(np.unique(y_train)) > 1 and len(y_train) >= 6 and min(np.bincount(y_train)) >= 6:
        smote = SMOTE(random_state=42)
        X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)
    else:
        X_train_smote, y_train_smote = X_train_scaled, y_train


    classes = y.unique()
    class_weights = compute_class_weight('balanced', classes=classes, y=y_train_smote)
    class_weight_dict = dict(zip(classes, class_weights))


    param_grid_simple = {
        'n_estimators': [100, 200, 500],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'max_features': ['sqrt', 'log2'],
        'bootstrap': [True, False]
    }


    rf = RandomForestClassifier(random_state=42, class_weight=class_weight_dict)
    rf_random_simple = RandomizedSearchCV(estimator=rf, param_distributions=param_grid_simple,
                                          n_iter=20, cv=3, verbose=1, random_state=42, n_jobs=-1)
    rf_random_simple.fit(X_train_smote, y_train_smote)
    best_rf_model_simple = rf_random_simple.best_estimator_


    y_pred_best_simple = best_rf_model_simple.predict(X_test_scaled)
    accuracy_best_simple = accuracy_score(y_test, y_pred_best_simple)
    classification_rep_best_simple = classification_report(y_test, y_pred_best_simple)
    conf_matrix_best_simple = confusion_matrix(y_test, y_pred_best_simple)


    valores=list_of_circuits[circuit]
    results.append({
        'Circuit': f'Circuit_{dictionary_circuits[valores]}',
        'Accuracy': accuracy_best_simple,
        'Classification Report': classification_rep_best_simple,
        'Confusion Matrix': conf_matrix_best_simple,
        'Feature Importance': best_rf_model_simple.feature_importances_,
        'Feature Names': X.columns
    })


    accuracy_summary.append(f"Accuracy Circuit_{dictionary_circuits[valores]}: {accuracy_best_simple * 100:.2f}%")
    
    print(f"\nResultados para Circuit_{dictionary_circuits[valores]}:")
    print("Precisión del modelo:", accuracy_best_simple)
    print("\nReporte de Clasificación:\n", classification_rep_best_simple)
    print("\nMatriz de Confusión:\n", conf_matrix_best_simple)

 
    important_features = pd.Series(best_rf_model_simple.feature_importances_, index=X.columns).sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    important_features.plot(kind='bar')
    plt.title(f'Importancia de Características - Circuit_{i+1}')
    plt.xlabel('Características')
    plt.ylabel('Importancia')
    plt.show()
    circuit+=1
    if(circuit==len(list_of_circuits)):
        circuit=0
results_df = pd.DataFrame(results)


    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

results = []
circuit = 0
accuracy_summary = []
valores = 0

for i, dataframe in enumerate(dataframes_circuits):
    # Filtrar y preparar datos
    df_filtered = dataframe[dataframe['target'] <= 20].dropna()
    X = df_filtered.drop(columns=['dob', 'date', 'driverref', 'circuitref', 'target', 'raceid'])
    y = df_filtered['target']

    # Dividir datos en entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

    # Escalar los datos
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Aplicar SMOTE para balancear las clases
    if len(np.unique(y_train)) > 1 and len(y_train) >= 6 and min(np.bincount(y_train)) >= 6:
        smote = SMOTE(random_state=42)
        X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)
    else:
        X_train_smote, y_train_smote = X_train_scaled, y_train

    num_classes = len(np.unique(y))  # Total de clases en los datos originales
    y_train_encoded = to_categorical(y_train_smote - 1, num_classes=num_classes)
    y_test_encoded = to_categorical(y_test - 1, num_classes=num_classes)

    print("Número de clases en y_train_encoded:", y_train_encoded.shape[1])
    print("Número de clases en y_test_encoded:", y_test_encoded.shape[1])

    output_units = num_classes  # Número total de clases

    model = Sequential([
        Dense(128, activation='relu', input_shape=(X_train_smote.shape[1],)),
        Dropout(0.3),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(output_units, activation='softmax')  # Asegúrate de que el número de salidas sea correcto
    ])

    # Compilar el modelo
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # Entrenar el modelo
    history = model.fit(X_train_smote, y_train_encoded,
                        epochs=50, batch_size=32,
                        validation_split=0.2, verbose=1)

    # Evaluar el modelo
    loss, accuracy = model.evaluate(X_test_scaled, y_test_encoded, verbose=0)
    y_pred = model.predict(X_test_scaled)
    y_pred_classes = y_pred.argmax(axis=1) + 1  # Regresamos a clases originales

    # Aquí usamos y_test directamente, porque ya es un vector de etiquetas
    classification_rep = classification_report(y_test, y_pred_classes)
    conf_matrix = confusion_matrix(y_test, y_pred_classes)

    valores = list_of_circuits[circuit]
    results.append({
        'Circuit': f'Circuit_{dictionary_circuits[valores]}',
        'Accuracy': accuracy,
        'Classification Report': classification_rep,
        'Confusion Matrix': conf_matrix
    })

    accuracy_summary.append(f"Accuracy Circuit_{dictionary_circuits[valores]}: {accuracy * 100:.2f}%")

    print(f"\nResultados para Circuit_{dictionary_circuits[valores]}:")
    print("Precisión del modelo:", accuracy)
    print("\nReporte de Clasificación:\n", classification_rep)
    print("\nMatriz de Confusión:\n", conf_matrix)

    # Graficar precisión y pérdida
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Pérdida de Entrenamiento')
    plt.plot(history.history['val_loss'], label='Pérdida de Validación')
    plt.legend()
    plt.title(f'Pérdida - Circuit_{i+1}')
    plt.xlabel('Época')
    plt.ylabel('Pérdida')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Precisión de Entrenamiento')
    plt.plot(history.history['val_accuracy'], label='Precisión de Validación')
    plt.legend()
    plt.title(f'Precisión - Circuit_{i+1}')
    plt.xlabel('Época')
    plt.ylabel('Precisión')

    plt.tight_layout()
    plt.show()

    circuit += 1
    if circuit == len(list_of_circuits):
        circuit = 0

results_df = pd.DataFrame(results)
```
